```python
# 1. Write a Python program to count the frequency of each word in a given string using a dictionary.

# Function to count word frequency
def count_word_frequency(input_string):
    # Convert the string to lowercase and split it into words
    words = input_string.lower().split()

    # Create an empty dictionary to store word frequencies
    word_freq = {}

    # Iterate through the words and count their occurrences
    for word in words:
        # Remove punctuation from the word
        word = word.strip(",.?!")
        # Update the dictionary
        word_freq[word] = word_freq.get(word, 0) + 1

    return word_freq

# Example input string
input_string = "Hello, world! Hello again. Welcome to the world of Python."

# Call the function and display the word frequencies
frequencies = count_word_frequency(input_string)
print("Word Frequencies:")
for word, count in frequencies.items():
    print(f"{word}: {count}")
```

```python
# 2. Write a Python program to read a text file and count the number of words in it.


# Function to count words in a file
def count_words_in_file(file_path):
    try:
        # Open the file in read mode
        with open(file_path, 'r') as file:
            # Read the content of the file
            content = file.read()
            # Split the content into words
            words = content.split()
            # Return the number of words
            return len(words)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None

# Example usage
file_path = "example.txt"  # Replace with the path to your text file
word_count = count_words_in_file(file_path)

if word_count is not None:
    print(f"The number of words in the file '{file_path}' is: {word_count}")
```

```python
#  5. Set up a local SQLite database and create a table with some sample data. Use Python to import
#  data from the SQL table into a data frame and perform any 5 operations on the data frame

import sqlite3
import pandas as pd

# Step 1: Set up SQLite database and create a table with sample data
conn = sqlite3.connect('local_database.db')  # Creates a new SQLite database file
cursor = conn.cursor()

# Create a table
cursor.execute('''
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER,
    hire_date TEXT
)
''')

# Insert some sample data into the table
cursor.executemany('''
INSERT INTO employees (name, department, salary, hire_date)
VALUES (?, ?, ?, ?)
''', [
    ('Alice', 'HR', 50000, '2015-06-01'),
    ('Bob', 'IT', 60000, '2016-08-15'),
    ('Charlie', 'Finance', 55000, '2017-03-20'),
    ('David', 'IT', 65000, '2018-01-10'),
    ('Eve', 'HR', 52000, '2019-09-25')
])

# Commit and close the connection
conn.commit()



# Step 2: Import data from SQLite into a Pandas DataFrame
query = "SELECT * FROM employees"
df = pd.read_sql_query(query, conn)

# Close the connection
conn.close()

# Display the DataFrame
print("DataFrame from SQLite Table:")
print(df)


# 1. Select specific columns (e.g., 'name' and 'salary')
print("\nSelected columns ('name' and 'salary'):")
print(df[['name', 'salary']])

# 2. Filter rows where salary > 55000
print("\nRows where salary > 55000:")
print(df[df['salary'] > 55000])

# 3. Sort the DataFrame by 'salary' in descending order
print("\nSorted by 'salary' in descending order:")
print(df.sort_values(by='salary', ascending=False))

# 4. Calculate the average salary
average_salary = df['salary'].mean()
print(f"\nAverage Salary: {average_salary:.2f}")

# 5. Group by department and calculate the average salary per department
print("\nAverage salary by department:")
print(df.groupby('department')['salary'].mean())
```

```python
# 6. Identify a website with tabular data (e.g., a Wikipedia table). Use BeautifulSoup to scrape the
#  data, store it in a Data Frame and perform any 5 operations on the data frame

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Define the URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Bicycle"

# Step 2: Fetch the webpage content
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Locate all tables with the 'wikitable' class
tables = soup.find_all("table", {"class": "wikitable"})

if not tables:
    print("No 'wikitable' tables found on the page.")
else:
    # Selecting the first table (you can choose other tables by adjusting the index)
    table = tables[0]

    # Step 4: Extract table headers
    headers = [header.text.strip() for header in table.find_all('th')]

    # Step 5: Extract table rows
    rows = []
    for row in table.find_all('tr')[1:]:  # Skip the header row
        cells = row.find_all(['td', 'th'])
        row_data = [cell.text.strip() for cell in cells]
        rows.append(row_data)

    # Step 6: Create a DataFrame
    df = pd.DataFrame(rows, columns=headers)

    # Step 7: Display the DataFrame
    print("DataFrame:")
    print(df.head())

    # --- Perform Operations ---

    # 1. Clean up column names (strip extra whitespace)
    df.columns = df.columns.str.strip()

    # 2. Drop rows with missing values
    df.dropna(inplace=True)

    # 3. Select specific columns (adjust to match table structure)
    if 'Type' in df.columns and 'Description' in df.columns:
        selected_columns = df[['Type', 'Description']]
        print("\nSelected Columns:")
        print(selected_columns.head())

    # 4. Sort the DataFrame by the first column (assume numeric or alphabetic)
    sorted_df = df.sort_values(by=df.columns[0], ascending=True)
    print("\nDataFrame sorted by first column:")
    print(sorted_df.head())

    # 5. Filter rows containing specific keywords in any column
    filter_keyword = 'bicycle'
    filtered_df = df[df.apply(lambda row: row.astype(str).str.contains(filter_keyword, case=False).any(), axis=1)]
    print(f"\nRows containing the keyword '{filter_keyword}':")
    print(filtered_df)

    # 6. Add a new column (e.g., length of text in the first column)
    df['Column_Length'] = df[df.columns[0]].apply(len)
    print("\nDataFrame with a new column (length of first column values):")
    print(df.head())

    # 7. Group data (if applicable)
    if 'Type' in df.columns:
        grouped = df.groupby('Type').size()
        print("\nGrouped data by 'Type':")
        print(grouped)

    # 8. Find basic statistics for numeric columns (if any exist)
    numeric_columns = df.select_dtypes(include=['number'])
    if not numeric_columns.empty:
        print("\nStatistics for numeric columns:")
        print(numeric_columns.describe())
    else:
        print("\nNo numeric columns found for statistics.")
```

```python
#  7. Identify a public API (e.g., OpenWeatherMap, CoinGecko). Use Python to fetch data from the
#  API, store it in a data frame and perform any 5 operations on the data frame.

import requests
import pandas as pd

# Step 1: Define the endpoint URL
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    "vs_currency": "usd",  # Currency for the data
    "order": "market_cap_desc",  # Order by market cap
    "per_page": 10,  # Number of results per page
    "page": 1,  # Page number
    "sparkline": False  # Exclude sparkline data
}

# Step 2: Fetch data from the API
response = requests.get(url, params=params)
data = response.json()

# Step 3: Create a DataFrame from the API response
crypto_df = pd.DataFrame(data)

# Display the DataFrame
print("Cryptocurrency Data:")
print(crypto_df.head())



# 1. Select specific columns
selected_columns = crypto_df[['id', 'symbol', 'current_price', 'market_cap', 'total_volume']]
print("\nSelected Columns:")
print(selected_columns.head())

# 2. Filter cryptocurrencies with a price greater than $1000
high_price_cryptos = crypto_df[crypto_df['current_price'] > 1000]
print("\nCryptocurrencies with a price greater than $1000:")
print(high_price_cryptos[['id', 'current_price']])

# 3. Sort the data by market cap in descending order
sorted_by_market_cap = crypto_df.sort_values(by='market_cap', ascending=False)
print("\nData sorted by Market Cap (Descending):")
print(sorted_by_market_cap[['id', 'market_cap']].head())

# 4. Find the average price of cryptocurrencies
average_price = crypto_df['current_price'].mean()
print(f"\nAverage price of cryptocurrencies: ${average_price:.2f}")

# 5. Group data by market cap range (High, Medium, Low)
def market_cap_category(market_cap):
    if market_cap > 1e10:
        return "High"
    elif market_cap > 1e9:
        return "Medium"
    else:
        return "Low"

crypto_df['market_cap_category'] = crypto_df['market_cap'].apply(market_cap_category)
grouped = crypto_df.groupby('market_cap_category').size()
print("\nCryptocurrencies grouped by Market Cap category:")
print(grouped)
```

```python
# 13. Perform a hypothesis test to determine if the mean of a sample differs significantly from a known
#  population mean.


import numpy as np
from scipy.stats import ttest_1samp

# Step 1: Define the sample data and population mean
sample_data = [85, 87, 90, 92, 88, 84, 89, 91, 93, 86]  # Sample test scores
population_mean = 88  # Known population mean

# Step 2: Perform a one-sample t-test
t_stat, p_value = ttest_1samp(sample_data, population_mean)

# Step 3: Display results
print("One-Sample t-Test Results:")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"Reject the null hypothesis (p < {alpha}). The sample mean differs significantly from the population mean.")
else:
    print(f"Fail to reject the null hypothesis (p >= {alpha}). No significant difference between sample and population mean.")
```

```python
#15.Implement a linear regression model using scikit-learn to predict house prices.

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

# Step 1: Load the dataset
data = pd.read_csv('housing.csv')  # Replace with the actual path

# Correct the variable name from 'df' to 'data'
data.isnull().sum()
data['total_bedrooms'] = data['total_bedrooms'].fillna(data['total_bedrooms'].mode()[0]) # Use 'data' instead of 'df'
print(data.isnull().sum())

# Correct the variable name from 'df' to 'data'
label_encoder = LabelEncoder()
data['ocean_proximity'] = label_encoder.fit_transform(data['ocean_proximity']) # Use 'data' instead of 'df'

# Step 2: Prepare the data
# Separate features (X) and target (y)
X = data.drop(columns=['median_house_value'])  # Replace 'Target_Column' with the name of the column containing house prices
y = data['median_house_value']

# Step 3: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 'data' is already dropped NA values above, no need to do it again
# data = data.dropna()

# Step 4: Initialize and train the Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Step 5: Make predictions on the test set
y_pred = lr_model.predict(X_test)

# Step 6: Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared metric

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")

# Optional: Display model coefficients and intercept
print("Coefficients:", lr_model.coef_)
print("Intercept:", lr_model.intercept_)
```

```python
# 16.Train and evaluate a Decision Tree Classifier on the following dataset

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Step 1: Load and preprocess the data
# Load the dataset
file_path = 'Admission_Predict.csv'
data = pd.read_csv(file_path)

# Drop the 'Serial No.' column as it is not informative
data = data.drop(columns=['Serial No.'])

# Create a binary target: 1 if 'Chance of Admit' >= 0.75, else 0
data['Admit'] = (data['Chance of Admit '] >= 0.75).astype(int)
data = data.drop(columns=['Chance of Admit '])

# Separate features (X) and target (y)
X = data.drop(columns=['Admit'])
y = data['Admit']

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Initialize and train the Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)

# Step 4: Make predictions on the test set
y_pred = dt_classifier.predict(X_test)

# Step 5: Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Display evaluation results
print(f"Accuracy: {accuracy:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# Step 6: Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(
    dt_classifier,
    filled=True,
    feature_names=X.columns,
    class_names=['Not Admitted', 'Admitted'],
    rounded=True
)
plt.title("Decision Tree Classifier for Admission Predict Dataset")
plt.show()
```
